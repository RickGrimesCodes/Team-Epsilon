```python
# Read the data into a Pandas DataFrame
games_info_df = pd.read_csv('Resources/games_info_clean.csv')
games_info_df.head()
```

```python
games_cleaned = games_info_df.loc[:, ['AppID', 'Name', 'Release date', 'Estimated owners',
                                       'DLC count', 'About the game', 'Windows', 'Mac', 'Linux',
                                       'Metacritic score', 'Publishers']]
games_cleaned
```

```python
games_cleaned.columns
```

```python
# Rename the blurb, launched_at, and deadline columns.
games_cleaned = games_cleaned.rename(columns={'Release date': 'Release Date', 'Estimated owners': 'Estimated Owners', 'DLC count': 'DLC Count',
                                             'About the game': 'Game Description', 'Metacritic score': 'Metacritic Score'})

games_cleaned.head()
```

```python
# Step 1: Combine  'Windows', 'Mac', and 'Linux' columns into a single column called 'Operating System(s)'
def combine_os(row):
    os_list = []
    if row['Windows']:
        os_list.append('Windows')
    if row['Mac']:
        os_list.append('Mac')
    if row['Linux']:
        os_list.append('Linux')
    return ', '.join(os_list)

# Apply the function to create the new column
games_cleaned['Operating System(s)'] = games_cleaned.apply(combine_os, axis=1)

# Step 2: Drop the original 'Windows', 'Mac', and 'Linux' columns
games_cleaned.drop(columns=['Windows', 'Mac', 'Linux'], inplace=True)

# Display the dataframe with the new comibined columns
games_cleaned.head(8)
```

```python
games_cleaned.info()
```

```python
# Create a new DataFrame for games that have a Metacritic score greater than 0
rated_games = games_cleaned[(games_cleaned["Metacritic Score"] > 0)]
rated_games.head()
```

```python
rated_games.info()
```

```python
# Export the rated_games dataframe as CSV files.
rated_games.to_csv("Resources/rated_games.csv", index=False)
```
